```python
#Open the file
file = open("/Users/samibialozynski/Documents/GitHub/PythonPracticum/Project3/colors.ppm")
```

```python
#Extract the PPM type, the width of the picture, the height of the picture, and the maximum color value
ppmOrRwv = file.readline()
secondline = file.readline()
width,height = secondline.split()
maxNumber = file.readline()

width = int(width)
height = int(height)
```

```python
#Read the rest of the file into a list 
numbers = []
for x in file :
    l = x.split()
    for a in l:
        numbers.append(a)
```

```python
if 'RWV' in ppmOrRwv:
    rwvfile = open("/Users/samibialozynski/Documents/GitHub/PythonPracticum/Project3/ppmtorwv.ppm", "w")
    #Add the PPM value, width, height, and max value
    #The width changes because the value is halved from the averages
    rwvfile.write("P3\n")
    rwvfile.write("%d %d\n" %(width, height))
    rwvfile.write(maxNumber)
    
    numbers = list(map(float, numbers))
    numbers = list(map(int, numbers))
    numbers = list(map(abs, numbers))

    #Place the rest of the numbers into the file
    i = 0
    while (i < len(numbers) - 3):
        rwvfile.writelines("%s\n" % numbers[i])
        i = i + 1
    rwvfile.close()
    
else:
    if rwvNumber == 0:
        ppmfile = open("/Users/samibialozynski/Documents/GitHub/PythonPracticum/Project3/regularppm.ppm", "w")
        #Add the PPM value, width, height, and max value
        #The width changes because the value is halved from the averages
        ppmfile.write("P3\n")
        ppmfile.write("%d %d\n" %(width, height))
        ppmfile.write(maxNumber)

        numbers = list(map(int, numbers))
        numbers = list(map(abs, numbers))

        #Place the rest of the numbers into the file
        i = 0
        while (i < len(numbers) - 3):
            ppmfile.writelines("%s\n" % numbers[i])
            i = i + 1

    else:
        for count in range(rwvNumber):
            #Average every 3 values with its pair
            #Split the values in the list numbers into red, green, blue values
            r = numbers[0 : len(numbers): 3]
            g = numbers[1 : len(numbers) : 3]
            b = numbers[2 : len(numbers): 3]

            #Map each of the color values together and convert them into integers
            r = list(map(int, r))
            g = list(map(int, g))
            b = list(map(int, b))

            #Split the colors values into 2 and average them
            avgR = [(a + b) / 2 for a, b in zip(r[::2], r[1::2])]
            avgG = [(a + b) / 2 for a, b in zip(g[::2], g[1::2])]
            avgB = [(a + b) / 2 for a, b in zip(b[::2], b[1::2])]

            #Put the averages into another map by value and convert them into integers again
            avgR = list(avgR)
            avgG = list(avgG)
            avgB = list(avgB)

            #Distribute the values back out pixel by pixel with the averages 
            newRGB = [y for x in zip(avgR, avgG, avgB) for y in x]        

            offsetR = [((a + b) / 2) - b for a, b in zip(r[::2], r[1::2])]
            offsetG = [((a + b) / 2) - b for a, b in zip(g[::2], g[1::2])]
            offsetB = [((a + b) / 2) - b for a, b in zip(b[::2], b[1::2])]

            offset = [y for x in zip(offsetR, offsetG, offsetB) for y in x]

            backtogether = [] 

            for h in range(height):
                for i in range(int(width/2 * 3)):
                    backtogether.append(newRGB[h * 128 * 3 + i])
                for i in range(int(width/2 * 3)):
                    backtogether.append(offset[h * 128 * 3 + i] + 127)

            red = []
            green = []
            blue = []
            
            offred = []
            offgreen = []
            offblue = []

            for i in range(int(height/2)):
                for j in range(int(width/2)):
                    redOne = i * int(width/2) * 3 * 2 + 3 * j
                    greenOne = redOne + 1
                    blueOne = redOne + 2
                    redTwo = i * int(width/2) * 3 * 2 + 3 * j + 3 * int(width/2)
                    greenTwo = redTwo + 1
                    blueTwo = redTwo + 2
                    averageReds = (newRGB[redOne] + newRGB[redTwo])/ 2
                    averageGreens = (newRGB[greenOne] + newRGB[greenTwo])/ 2
                    averageBlues = (newRGB[blueOne] + newRGB[blueTwo])/ 2
                    
                    red.append(averageReds)
                    green.append(averageGreens)
                    blue.append(averageBlues)
                    
                    offredd = averageReds - newRGB[redTwo]
                    offgreenn = averageGreens- newRGB[greenTwo]
                    offbluee = averageBlues - newRGB[blueTwo]
                    
                    offred.append(offredd)
                    offgreen.append(offgreenn)
                    offblue.append(offbluee)
                    
                    
                    
                    

            newest = [y for x in zip(red, green, blue) for y in x]

            offnewest = [y for x in zip(offred, offgreen, offblue) for y in x]

            print(offnewest)
            
            alltogether = []

            for h in range(int(height/2)):
                for i in range(int(width/2 * 3)):
                    alltogether.append(newest[h * 128 * 3 + i])
                for i in range(int(width/2 * 3)):
                    alltogether.append(offnewest[h * 128 * 3 + i] + 127)

            rvwFile = open("/Users/samibialozynski/Documents/GitHub/PythonPracticum/Project3/waves.rwv", "w")
            rvwFile.write("RWV%d\n" % rwvNumber)
            rvwFile.write("%d %d\n" %(width, height))
            rvwFile.write(maxNumber)

            z = 0
            while (z < len(alltogether) - 3):
                rvwFile.writelines("%s\n" % alltogether[z])
                z = z + 1

            alltogether = list(map(int, alltogether))
            #alltogether = list(map(abs, alltogether))
```
