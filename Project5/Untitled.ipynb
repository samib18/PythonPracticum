```python
#Open the file
file = open("/Users/samibialozynski/Documents/GitHub/PythonPracticum/Project/balls.ppm")

#Extract the PPM type, the width of the picture, the height of the picture, and the maximum color value
ppmOrRwv = file.readline()
secondline = file.readline()
width,height = secondline.split()
maxNumber = file.readline()

width = int(width)
height = int(height)

#Read the rest of the file into a list 
numbers = []
for x in file :
    l = x.split()
    for a in l:
        numbers.append(a)

            
#Average every 3 values with its pair
#Split the values in the list numbers into red, green, blue values
r = numbers[0 : len(numbers): 3]
g = numbers[1 : len(numbers) : 3]
b = numbers[2 : len(numbers): 3]

#Map each of the color values together and convert them into integers
r = list(map(int, r))
g = list(map(int, g))
b = list(map(int, b))

#Split the colors values into 2 and average them
avgR = [(a + b) / 2 for a, b in zip(r[::2], r[1::2])]
avgG = [(a + b) / 2 for a, b in zip(g[::2], g[1::2])]
avgB = [(a + b) / 2 for a, b in zip(b[::2], b[1::2])]

#Put the averages into another map by value and convert them into integers again
avgR = list(avgR)
avgG = list(avgG)
avgB = list(avgB)

#Distribute the values back out pixel by pixel with the averages 
newRGB = [y for x in zip(avgR, avgG, avgB) for y in x]        

offsetR = [b - ((a + b) / 2)  for a, b in zip(r[::2], r[1::2])]
offsetG = [b - ((a + b) / 2) for a, b in zip(g[::2], g[1::2])]
offsetB = [b - ((a + b) / 2) for a, b in zip(b[::2], b[1::2])]

offset = [y for x in zip(offsetR, offsetG, offsetB) for y in x]

backtogether = []

for h in range(int(height)):
    for i in range(int(width/2 * 3)):
        backtogether.append(newRGB[h * int(width/2 * 3) +  i])
    for i in range(int(width/2 * 3)):
        backtogether.append(offset[h * int(width/2 * 3) + i] + 127.5)

backtogether = list(map(int, backtogether))
backtogether = list(map(abs, backtogether))

red = []
green = []
blue = []

offred = []
offgreen = []
offblue = []
alltogether = []

for i in range(int(height/ 2)):
    for j in range(int(width)):
        redOne = i * int(width) * 3 * 2 + 3 * j
        greenOne = redOne + 1
        blueOne = redOne + 2
        redTwo = i * int(width) * 3 * 2 + 3 * j + 3 * int(width)
        greenTwo = redTwo + 1
        blueTwo = redTwo + 2

        averageReds = (backtogether[redOne] + backtogether[redTwo]) / 2
        averageGreens = (backtogether[greenOne] + backtogether[greenTwo])/ 2
        averageBlues = (backtogether[blueOne] + backtogether[blueTwo])/ 2

        red.append(averageReds)
        green.append(averageGreens)
        blue.append(averageBlues)

        offredd = backtogether[redOne] - averageReds
        offgreenn = backtogether[greenOne] - averageGreens
        offbluee = backtogether[blueOne] - averageBlues 

        offred.append(offredd)
        offgreen.append(offgreenn)
        offblue.append(offbluee)                   

newest = [y for x in zip(red, green, blue) for y in x]
offnewest = [y for x in zip(offred, offgreen, offblue) for y in x]

offnewest = [x + 127.5 for x in offnewest]

alltogether = newest + offnewest

alltogether = list(map(int, alltogether))
alltogether = list(map(abs, alltogether))

rwvNumber = rwvNumber - 1

rvwFile = open("/Users/samibialozynski/Documents/GitHub/PythonPracticum/Project3/new.rwv", "w")
#rvwFile.write("RWV%d\n" % rwvNumber)
rvwFile.write("P3\n")
rvwFile.write("%d %d\n" %(width, height))
rvwFile.write(maxNumber)

z = 0
while (z < len(alltogether)):
    rvwFile.writelines("%s\n" % alltogether[z])
    z = z + 1

rvwFile.close()
file.close()
```
