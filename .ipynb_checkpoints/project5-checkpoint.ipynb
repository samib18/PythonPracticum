```python
#Create RLE format
#Open the file
rwvFile = open("/Users/samibialozynski/Documents/GitHub/PythonPracticum/Project5/rwv1.rwv")

#Extract the RWV type, the width of the picture, the height of the picture, and the maximum color value
rwv = rwvFile.readline()
secondline = rwvFile.readline()
width,height = secondline.split()
maxNumber = rwvFile.readline()

width = int(width)
height = int(height)

#Read the rest of the file into a list 
numbers = []
for x in rwvFile:
    l = x.split()
    for a in l:
        numbers.append(a)

        
threshold = float(1.5)
numbers = list(map(float, numbers))

setthres = [] 
count = 0

for num in numbers:
    if abs(num) < threshold:
        setthres.append(0)
    else:
        setthres.append(num)
        
print(setthres.count(0))

rle = []
count0 = 0

for zero in setthres:
    if zero == float(0):
        if count0 == 0 or count0 != 0:
            count0 += 1
    else:
        if count0 != 0:
            rle.append(int(0))
            rle.append(count0)
            rle.append(zero)
            count0 = 0
        else:
            rle.append(zero)
if count0 != 0:
    rle.append(int(0))
    rle.append(count0)  
    
print(rle.count(0))

rleFile = open("/Users/samibialozynski/Documents/GitHub/PythonPracticum/Project5/colorsrwv1rle.rwv", "w")
rleFile.write("RWV1\n")
rleFile.write("%d %d\n" %(width, height))
rleFile.write(maxNumber)

z = 0
while (z < len(rle)):
    rleFile.writelines("%s\n" % rle[z])
    z = z + 1

rleFile.close()
rwvFile.close()
```

```python
#Open the file
rleFile = open("/Users/samibialozynski/Documents/GitHub/PythonPracticum/Project5/colorsrwv1rle.rwv")

#Extract the RWV type, the width of the picture, the height of the picture, and the maximum color value
rwv = rleFile.readline()
secondline = rleFile.readline()
width,height = secondline.split()
maxNumber = rleFile.readline()

width = int(width)
height = int(height)

#Read the rest of the file into a list 
numbers = []
for x in rleFile:
    l = x.split()
    for a in l:
        numbers.append(a)
        
together = []
t = 0

while t < len(numbers):
    if numbers[t] == '0': 
        num = int(numbers[t + 1])
        for i in range(num):
            together.append('0')
        t += 2
    else:
        together.append(numbers[t])
        t += 1
        
print(together.count("0"))

offset = int(height // 2  * width * 3)
ysh = [] 
xsh = [] 
i = 0

for z in range(height//2): 
    for off in range(width * 3): 
        y = numbers[i] - numbers[i + offset] + 127.5 
        ysh.append(y) 
        x = 2 * int(numbers[i]) - y 
        xsh.append(x) 
        i += 1
        
together2 = [] 

for h in range(height//2): 
    for x in range(width * 3): 
        together2.append(xsh[x + width * 3 * h]) 
    for y in range(width * 3): 
        together2.append(ysh[y + width * 3 *h])
        
newoffset = width// 2 * 3 
xs = []
ys = []
for t in range(height): 
    for l in range(width // 2 * 3): 
        r = together2[l] - numbers[l + newoffset] + 127.5
        xs.append(r)
        f = together2[1] - numbers[l + newoffset]
        ys.append(f)
        

rleFile = open("/Users/samibialozynski/Documents/GitHub/PythonPracticum/Project5/back.ppm", "w")
rleFile.write("RWV1\n")
rleFile.write("%d %d\n" %(width, height))
rleFile.write(maxNumber)

z = 0
while (z < len(together)):
    rleFile.writelines("%s\n" % together[z])
    z = z + 1

rleFile.close()
rwvFile.close()
```
